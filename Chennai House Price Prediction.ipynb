```python
ChennaiHousing['PARK_FACIL']=ChennaiHousing['PARK_FACIL'].map({'Yes':1,'No':0})
```

```python
ChennaiHousing['BUILDTYPE'].unique()
```

```python
ChennaiHousing=ChennaiHousing.dropna()
```

```python
X=ChennaiHousing[['INT_SQFT','N_ROOM','N_BEDROOM','REG_FEE']]
y=ChennaiHousing['SALES_PRICE']
```

```python
X.isnull().values.any()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm=LinearRegression()
```

```python
lm.fit(X_train,y_train)
```

```python
print(lm.intercept_)
```

```python
coeff_df = pd.DataFrame(lm.coef_,X.columns,columns=['Coefficient'])
coeff_df
```

It seems that the number of rooms and registrations fees makes a impact on the final prices.

```python
predictions = lm.predict(X_test)
```

```python
plt.scatter(y_test,predictions)
```

```python
sns.displot((y_test-predictions),bins=50,kde=True);
```

```python
from sklearn import metrics
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('R2- SCORE:', metrics.r2_score(y_test,predictions))
```

Since we get Mean absolute error to 1263547 approx, it can be said that the buyer can negotiate the house price upto 12.6 Lacs

Decision tree model

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dt=DecisionTreeRegressor()
```

```python
dt.fit(X_train,y_train)
```

```python
y_pred=dt.predict(X_test)
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, y_pred))
print('MSE:', metrics.mean_squared_error(y_test, y_pred))
print('R2- SCORE:', metrics.r2_score(y_test,y_pred))
```

Seems like the R2 score is very poor compared to Linear Regressor model

Hence we choose the linear regression model to predict the final price
