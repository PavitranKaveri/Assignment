```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
t1=np.random.randint(1,50,size=50)
t2=np.random.randint(1,25,size=50)
y1=np.random.randint(1,50,size=50)
y2=np.random.randint(1,25,size=50)
```

```python
plt.figure(figsize=(10,6))
f1=plt.subplot2grid((5,6),(0,0),rowspan=1,colspan=3)
f2=plt.subplot2grid((5,6),(0,3),rowspan=1,colspan=1)
f3=plt.subplot2grid((5,6),(0,4),rowspan=1,colspan=2)
f4=plt.subplot2grid((5,6),(1,0),rowspan=3,colspan=1)
f5=plt.subplot2grid((5,6),(1,1),rowspan=1,colspan=2)
f6=plt.subplot2grid((5,6),(1,3),rowspan=1,colspan=3)
f7=plt.subplot2grid((5,6),(2,1),rowspan=1,colspan=1)
f8=plt.subplot2grid((5,6),(3,1),rowspan=1,colspan=1)
f9=plt.subplot2grid((5,6),(4,0),rowspan=1,colspan=2)
f10=plt.subplot2grid((5,6),(2,2),rowspan=3,colspan=2)
f11=plt.subplot2grid((5,6),(2,4),rowspan=3,colspan=1)
f12=plt.subplot2grid((5,6),(2,5),rowspan=3,colspan=1)

f1.barh(t1,y1,color='red')
f2.scatter(t2,y2,color='lime')
f3.bar(y1,y2,color='indigo')
f4.plot(t2,y2,color='crimson')
f5.barh(t1,y1,color='mediumvioletred')
f6.scatter(t2,y2,color='pink')
f7.bar(y1,y2,color='darkcyan')
f8.plot(t2,y2,color='darkolivegreen')
f9.barh(t1,y1,color='green')
f10.scatter(t2,y2,color='coral')
f11.bar(y1,y2,color='maroon')
f12.plot(t2,y2,color='gold')
plt.tight_layout()
```
